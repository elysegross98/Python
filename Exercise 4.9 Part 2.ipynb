## Creating Bar Charts

## Line Chart and Sampling

```python
#We need a randomized set of rows

np.random.seed(4)
dev = np.random.rand(len(df_merged)) <= 0.7
```

```python
#Store 70% of the sample in dataframe big
big = df_merged[dev]

#Store 30% of the sample in dataframe small
small = df_merged[~dev]
```

```python
#Checking to make sure lengths are the same 

len(df_merged)
```

```python
len(big) + len(small)
```

```python
#Reducing size more to speed up the process
#Only including necessary columns

df_2 = small[['orders_day_of_week', 'prices']]
```

```python
#Ensuring the above worked

df_2.head()
```

```python
#Creating a linechart with days of week of orders and prices using new sample of data

line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```

```python
#Export line chart outside jupyter notebook

line.figure.savefig(os.path.join(path1, '04 Analysis','Visualizations', 'line_days_prices.png'))
```

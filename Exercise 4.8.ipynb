## Step 4. Create the Loyalty Flag

Creating a loyalty flag for the data
A loyal customer is one that comes back time and time again.

#### How we will determine loyalty:

Customer has > 40 max orders - "Loyal customer"
Cutsomer has 10 > 40 max orders - "Regular customer"
Customer has < 10 orders - "New customer"

The criteria for flags is often provided by the client

```python
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merged.size
```

```python
# To allow us to see as many rows as we want
pd.options.display.max_rows = None
```

```python
#Viewing the new dataframe 
df_ords_prods_merged.head(100)
```

```python
# Creating the loyalty labels in the column

df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Viewing only the columns we want to see with the 
    #new loyalty labels

df_ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## Step 5. Are the prices of products purchased by loyal customers different from those purchased by regular or new customers?

```python
# Creating a smaller dataframe so it's easier to work with

df_prices = df_ords_prods_merged[['user_id', 'loyalty_flag', 'order_number', 'prices']]
                      
```

```python
# Confirming we have the desired columns

df_prices.head()
```

```python
#Groupoing by loyalty flag to determine whether higher 
    #loyalty customers spending.
# Including muliple summary statistics to "check the 
    #basic statistics of product prices".
df_prices.groupby('loyalty_flag').agg({'prices': ['sum', 'mean', 'min', 'max']})
```

Though the loyal customers have spent more over time (as we see in sum), we see they have slightly lower average spending accross products (as we see in mean).

## Step 6. Targeting different spenders in marketing campaigns.

### Create a spending flag for each user based on the average price accross all their orders.

#### Criteria:
   
mean of prices < 10 - "Low spender"
mean of prices >= 10 - "High spender"

```python
df_ords_prods_merged['price_means'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merged.head()
```

```python
# Creating the spending flags in the column

df_ords_prods_merged.loc[df_ords_prods_merged['price_means'] < 10, 'spender_type'] = 'Low Spender'
df_ords_prods_merged.loc[(df_ords_prods_merged['price_means'] >= 10), 'spender_type'] = 'High Spender'
```

```python
# Viewing only the columsn we want to see with new 
    #spending labels.

df_ords_prods_merged[['user_id', 'price_means', 'spender_type']].head(60)
```

## Step 7. Frequent vs. Non-frequent customers

#### Criteria:
   
median of days_since_prior_order > 20 - "Non-frequent customer"
median of days_since_prior_order 10 > 20 - "Regular customer"
median of days_since_prior_order <= 10 - "Frequent customer"

```python
df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Viewing data
df_ords_prods_merged.head(100)
```

```python
# Creating the loyalty labels in the column

df_ords_prods_merged.loc[(df_ords_prods_merged['median_days_since_prior_order'] > 20)  & (df_ords_prods_merged['median_days_since_prior_order'] == 'NaN'), 'spending_flag'] = 'Non-frequent customer'
df_ords_prods_merged.loc[(df_ords_prods_merged['median_days_since_prior_order'] <= 20) & (df_ords_prods_merged['median_days_since_prior_order'] > 10), 'spending_flag'] = 'Regular customer'
df_ords_prods_merged.loc[(df_ords_prods_merged['median_days_since_prior_order'] <= 10), 'spending_flag'] = 'Frequent customer'
```

```python
# Viewing only the columns we want to see with the 
    #new spending flags.

df_ords_prods_merged[['user_id', 'spending_flag', 'median_days_since_prior_order']].head(60)
```

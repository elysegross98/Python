## Sort Products into Price Ranges with loc( )

```python
#Using loc instead of an if construct statement
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

## For loop for busiest day

```python
df_ords_prods_merged.head()
```

```python
#Printing frequency to know which values appear most often in the column
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a for loop to see which day is the busiest

result = []

for value in df_ords_prods_merged['orders_day_of_week']:
        if value == 0:
            result.append("Busiest day")
        elif value == 6:
            result.append("Least busy")
        else: 
            result.append("Regular busy")
```

```python
#Print result list

result
```

```python
#Combine the list we created with the orders products df

df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

## Step 2. New Busiest days column

```python
# Create a for loop to see which day is the busiest

result_new = []

for value in df_ords_prods_merged['orders_day_of_week']:
        if value == 0:
            result_new.append("Busiest day")
        elif value ==1:
            result_new.append("Second Busiest day")
        elif value == 6:
            result_new.append("Least busy")
        elif value == 5:
            result_new.append("Second Least Busy")
        else: 
            result_new.append("Regular busy")
```

```python
#Viewing new results
result_new
```

```python
#Adding this as a new column to the dataframe
df_ords_prods_merged['busiest_days'] = result_new
```

```python
#Viewing the new column to check the values for accuracy
df_ords_prods_merged
```

The values in this new column look good. They correspond with the busiest and least busy days of the previous busiest_day column. We see there is more detail now, though, with the 'second busiest day' and 'second least busy'.

## Step 4. Busiest Period of the Day

```python
#Printing frequency to know which values appear most often in the column
df_ords_prods_merged['order_hour'].value_counts(dropna = False)
```

```python
#Creating a for loop to make a list of business of periods

result_period = []

for value in df_ords_prods_merged['order_hour']:
    if value in [4, 3, 5, 2, 1, 0, 6, 7]:
        result_period.append("Fewest orders")
    elif value in [23, 8, 22, 21, 9, 20, 19, 10]:
        result_period.append("Average orders")
    elif value in [18, 11, 12, 17, 13, 14, 15, 16]:
        result_period.append("Most orders")
```

```python
#Adding this as a new column to the dataframe
df_ords_prods_merged['busiest_period_of_day'] = result_period
```

```python
#Viewing the new column
df_ords_prods_merged.head()
```

## Step 5. Frequency for new column

```python
#Printing the freuquency of this new column
df_ords_prods_merged['busiest_period_of_day'].value_counts()
```
